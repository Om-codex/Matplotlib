```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# print(plt.style.available)
plt.style.use('seaborn-v0_8')

x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]
colors = [7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]
sizes = [209,486,381,255,191,315,185,228,174,538,239,394,399,153,273,293,436,501,397,539]

plt.scatter(x,y,s= sizes,c= colors,cmap='rainbow' ,edgecolor='black',linewidth=1,alpha=0.75)
cbar= plt.colorbar()
cbar.set_label('Satisfaction Level',fontsize=15,fontweight='bold')

plt.tight_layout()
plt.figure(figsize=(8,6))
plt.show()
```

```python
data = pd.read_csv('yt_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
plt.figure(figsize=(10,6))

plt.scatter(view_count,likes,s=70,cmap='autumn',edgecolor = 'black',c=ratio,linewidth=1,alpha = 0.75)
plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label('Like Dislike Ratio')
plt.title('Trending Youtube Videos',fontweight='bold')
plt.xlabel('View Count',fontweight='bold')
plt.ylabel('Total Likes',fontweight='bold')
plt.tight_layout()
plt.show()
```
