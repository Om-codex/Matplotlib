```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use('fivethirtyeight')

slices = [60,20,20,50] #Values
labels = ['Sixty','Twenty','Twenty','Fifty'] #giving names to respective values
colors = ['Cyan','seagreen','Magenta','coral'] # applying different colors
plt.pie(slices,labels=labels,colors = colors,wedgeprops={'edgecolor':'black'})#for black outline

plt.title('Learning Pie Chart')
plt.tight_layout()
plt.show()
```

# #Poplar Languages Dataset Piechart

```python
plt.style.use('fivethirtyeight')

slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0] #To seperate a specific data from the pie chart that is to offset it

plt.pie(slices,labels = labels, 
        wedgeprops = {'edgecolor':'black'},#wedgeprops gives the outline color
        explode = explode,                 #explode seperates it 
        shadow = True,                     #shadow gives a shadow to the pie
        startangle=120,                    #startangle rotates pie based on the angle
        autopct = '%1.1f%%')               #autopct gives percentage values     





plt.title('Popular Langauages')
plt.tight_layout()
plt.show()
```
